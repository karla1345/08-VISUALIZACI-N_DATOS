+ `education_level`: El nivel más alto de educación alcanzado por el individuo, como 'Escuela secundaria', 'Asociado', 'Licenciatura', 'Maestría' o 'Doctorado'.

# 2. Crea un grafico de barras en donde se muestren la cantidad de personas registradas por `occupation`.

```python
a=prestamo['occupation'].value_counts()
plt.figure(figsize=(10,6))  #tamaño de gráfico
plt.bar(a.index,a.values,color='red')

#añadir título y  etiquetas a los ejes.
plt.title('Profesiones de los clientes')
plt.xlabel('Profeciones')#poner etiqueta a eje x
plt.ylabel('Número de Clientes')#poner etiqueta a eje y
plt.xticks(rotation=90)
plt.show()#mostrar Gráfico
```

# 4.Crea un grafico circular en donde se muestre la proporcion de Hombres y mujeres de acuerdo a la columna `gender`

```python
prestamo['gender'].value_counts() # Calcula el conteo de valores únicos en la columna 'gender' del DataFrame prestamo
```

```python
a=prestamo['gender'].value_counts()
plt.figure(figsize=(10,6))  #tamaño de gráfico
plt.pie(a.values, labels=a.index, colors=['red', 'blue'], autopct='%1.1f%%')

#añadir título y  etiquetas a los ejes.
plt.title('Proporción de hombres y mujeres')
plt.show()#mostrar Gráfico
```

# 5. Crea un histograma con la columna `age`

```python
prestamo['age'].value_counts() # Calcula el conteo de valores únicos en la columna 'age' del DataFrame prestamo
```

```python
a=prestamo['age'].value_counts()
plt.figure(figsize=(10, 6))  # Tamaño del gráfico
plt.hist(prestamo['age'], bins=10, color='red', edgecolor='black')  # Crear histograma

# Añadir título y etiquetas a los ejes.
plt.title('Distribución de Edad')
plt.xlabel('Edad')  # Poner etiqueta al eje xÑ
plt.ylabel('Número de Clientes')  # Poner etiqueta al eje y
plt.xticks(rotation=90)  # Rotar las etiquetas del eje x si es necesario
plt.show()  # Mostrar gráfico
```
